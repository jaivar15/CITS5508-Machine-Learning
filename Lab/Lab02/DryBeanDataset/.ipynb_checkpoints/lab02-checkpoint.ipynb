## Section 2:  Support Vector Machines for classification of Dry Beans dataset

### Subsection 2.1:  Preprocessing data

```python
from sklearn.model_selection import train_test_split
```

Partition the dataset into X (input features) and y (target labels). X consists of every column in dataframe except for 'Class' column, while y consists of only the 'Class column'.

```python
X = dry_beans.drop(columns=['Class'])
y = dry_beans['Class']
```

Partition X and y datasets into training and test sets for both X and y. Test dataset should be 20% of total dataset. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape  # Output dimensions of X training set
```

```python
X_test.shape   # Output dimensions of X test set
```

For scaling, use MinMaxScaler instead of StandardScaler because several of the input features do not appear to be normally distributed.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

### Subsection 2.2:  Training and testing model 

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

Gridsearch description:
- Investigate the 'kernel,'C' and 'gamma' hyperparameters. 
- Have one grid for testing different kernel functions, and another grid for testing different values of 'C' and 'gamma' hyperparameters for the 'rbf' kernel function. 
- The 'C' and 'gamma' values tested include the default value and values that are increased or decreased by a factor of 10 relative to the default value.

Parameter descriptions:
- Kernel: Refers to the type of kernel function used, which is used to implement the kernel trick.
- C: A regularisation hyperparameter which controls the extent of point misclassification.
- Gamma:  A regularisation hyperparamter which controls the radius of influence of sample points near the decision boundary.

```python
param_grid_svc = [{'kernel':['rbf','poly','sigmoid']},{'kernel':['rbf'],'C':[0.1,1,10],'gamma':['auto',0.00625,0.625]}]
```

```python
grid_svc = GridSearchCV(SVC(),param_grid_svc)
```

```python
grid_svc.fit(X_train,y_train)
```

```python
grid_svc.best_params_
```

```python
y_pred_svc = grid_svc.predict(X_test)
```

### Subsection 2.3:  Evaluating model

Generate a confusion matrix plot and a classification report containing key metrics.

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
```

```python
plot_confusion_matrix(grid_svc, X_test, y_test)
```

```python
print(classification_report(y_test,y_pred_svc))
```

## Section 3:  Stochastic Gradient Descent for classification of Dry Beans dataset

### Subsection 3.1:  Training and testing model

```python
from sklearn.linear_model import SGDClassifier
```

Gridsearch description:
- In the grid search for SGD, investigate the 'penalty', 'alpha', 'learning_rate' and 'eta0' hyperparameters. 
- Have one grid for testing the 'penalty' and 'alpha'. hyperparameters
- Have another grid for testing the 'learning_rate' hyperparameter for the 'constant' and 'invscaling' options and an 'eta0' value of 1.

Hyperparameter descriptions:
- Penalty:  Refers to the type of regularisation used (either l1 or l2).
- Alpha:  A coefficient that is multiplied with regularisation term. The higher the alpha, the stronger the regularisation.
- Learning rate: Refers to the learning rate schedule. Different types include 'constant', 'optimal', 'invscaling'.
- Eta0:  Refers to the initial learning rate used in the 'constant', 'adaptive' and 'invscaling' learning rate schedules.

```python
param_grid_sgd = [{'penalty':['l2','l1'],'alpha':[0.00001,0.0001,0.001],'learning_rate':['optimal']},{'learning_rate':['constant','invscaling'],'eta0':[1]}]
```

```python
grid_sgd = GridSearchCV(SGDClassifier(),param_grid_sgd)
```

```python
grid_sgd.fit(X_train,y_train)
```

```python
grid_sgd.best_params_
```

```python
y_pred_sgd = grid_sgd.predict(X_test)
```

### Subsection 3.1:  Evaluating model

Generate a confusion matrix plot and a classification report containing key metrics.

```python
plot_confusion_matrix(grid_sgd, X_test, y_test)
```

```python
print(classification_report(y_test,y_pred_sgd))
```
